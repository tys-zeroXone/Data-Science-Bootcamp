# Soal 1
Buatlah sebuah fungsi untuk mendapatkan nilai terbesar dari input 3 (tiga) buah bilangan

```python
def cari_maksimum(bilangan1, bilangan2, bilangan3):
  if bilangan1 >= bilangan2 and bilangan1 >= bilangan3:
    return bilangan1
  elif bilangan2 >= bilangan1 and bilangan2 >= bilangan3:
    return bilangan2
  else:
    return bilangan3


#pengujian fungsi
# print(f"Nilai terbesar dari (5,5,8): {cari_maksimum(5,12,8)}")
# print(f"Nilai terbesar dari 13,5,2): {cari_maksimum(13,5,2)}")
```

# Soal 2
Buatlah sebuah fungsi untuk menentukan nilai rata-rata dari input sejumlah n bilangan dalam sebuah list

```python
def cari_rata2(input):

  total = 0

  for item in input:
    total += item

  return total/len(input)


# pengujian fungsi
# print(cari_rata2([10,20,30,40,50]))
```

# Soal 3
Buatlah sebuah fungsi yang menghitung banyaknya bilangan genap dan bilangan ganjil dari sejumlah elemen dari dalam sebuah list.

Output berupa sebuah tuple dengan 2 elemen:
*   elemen pertama adalah banyaknya bilangan genap
*   elemen kedua adalah banyaknya bilangan ganjil


```python
def jumlah_ganjil_genap(input):

  jumlah_genap = 0
  jumlah_ganjil = 0

  for item in input:
    if item%2 == 0:
      jumlah_genap +=1
    else:
      jumlah_ganjil +=1

  return (jumlah_genap, jumlah_ganjil)


#pengujian fungsi
# input_list = [1, 2, 3, 4, 5, 6, 7]
# print(f"Input: {input_list}")
# print(f"Output (genap, ganjil): {jumlah_ganjil_genap(input_list)}")
```

# Soal 4
Buatlah sebuah fungsi yang mengubah susunan huruf dalam string tersebut supaya hurup kecil muncul terlebih dahulu diikuti huruf kapital

```python
def ubah_urutan(input):

  list_lower = ''
  list_upper = ''

  for item in input:
    if item.islower() == True:
      list_lower += item
    else:
      list_upper += item

  return list_lower + list_upper


# pengujian fungsi
# input_text = "DigitaLSkolA"
# print(f"Input: {input_text}")
# print(f"Output: {ubah_urutan(input_text)}")
```

# Soal 5
Buatlah fungsi yang menghitung banyaknya kemunculan setiap karakter (huruf/angka/symbol) dalam kata tersebut. Hasil kembalian harus berupa dictionary

```python
def hitung_karakter(input):

  output = {}

  for item in input:
    if output.get('item') == None:
      output.update({item: 1})
    else:
      output.update({item: output.get('item')+1})

  return output

#pengujian fungsi
# input_text = "data science"
# print(f"Input: {input_text}")
# print(f"Output: {hitung_karakter(input_text)}")
# print()
```
