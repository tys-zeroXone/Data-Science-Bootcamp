# Soal no. 1 & 2

```python
# Set style untuk visualisasi yang lebih menarik
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)

print("=" * 80)
print("ANALISIS VISUALISASI DATA CREDIT CARD CUSTOMERS")
print("=" * 80)
```

## 1.1 ATTRITION FLAG - PIE CHART

```python
plt.figure(figsize=(10, 6))
attrition_counts = df['Attrition_Flag'].value_counts()
colors = ['#66b3ff', '#ff9999']
explode = (0.05, 0)

plt.pie(attrition_counts, labels=attrition_counts.index, autopct='%1.1f%%',
        startangle=90, colors=colors, explode=explode,
        shadow=True, textprops={'fontsize': 12, 'weight': 'bold'})
plt.title('Distribusi Customer Attrition Status', fontsize=16, weight='bold', pad=20)
plt.axis('equal')
plt.tight_layout()
plt.savefig('1_attrition_pie.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n1. ATTRITION FLAG (Pie Chart)")
print("-" * 80)
print(f"Existing Customer: {attrition_counts['Existing Customer']} ({attrition_counts['Existing Customer']/len(df)*100:.1f}%)")
print(f"Attrited Customer: {attrition_counts['Attrited Customer']} ({attrition_counts['Attrited Customer']/len(df)*100:.1f}%)")
```

### INSIGHT:
- Mayoritas customer (83.9%) masih aktif menggunakan layanan kartu kredit
- Tingkat churn sebesar 16.1% menunjukkan ada potensi kehilangan customer
- Bank perlu fokus pada strategi retensi untuk mengurangi attrition rate

## 1.2 CUSTOMER AGE - DONUT CHART

```python
plt.figure(figsize=(10, 6))
age_bins = [0, 30, 40, 50, 60, 100]
age_labels = ['<30', '30-40', '40-50', '50-60', '60+']
df['age_group'] = pd.cut(df['Customer_Age'], bins=age_bins, labels=age_labels)
age_counts = df['age_group'].value_counts().sort_index()

colors_donut = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc']
wedges, texts, autotexts = plt.pie(age_counts, labels=age_counts.index, autopct='%1.1f%%',
                                     startangle=90, colors=colors_donut,
                                     pctdistance=0.85, textprops={'fontsize': 11, 'weight': 'bold'})

# Membuat donut chart dengan menambahkan circle di tengah
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Distribusi Kelompok Usia Customer', fontsize=16, weight='bold', pad=20)
plt.axis('equal')
plt.tight_layout()
plt.savefig('2_age_donut.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n2. CUSTOMER AGE (Donut Chart)")
print("-" * 80)
for label, count in age_counts.items():
    print(f"{label} tahun: {count} ({count/len(df)*100:.1f}%)")
```

### INSIGHT:
- Mayoritas customer berada di rentang usia 40-50 tahun (33.8%) dan 50-60 tahun (29.7%)
- Customer berusia muda (<30 tahun) hanya 7.3%, menunjukkan potensi pasar muda belum tergarap
- Fokus marketing bisa ditargetkan ke generasi muda untuk meningkatkan customer base

## 1.3 GENDER - VERTICAL BAR CHART

```python
plt.figure(figsize=(10, 6))
gender_counts = df['Gender'].value_counts()
colors_gender = ['#ff6b6b', '#4ecdc4']

bars = plt.bar(gender_counts.index, gender_counts.values, color=colors_gender,
               edgecolor='black', linewidth=1.5, alpha=0.8)

# Menambahkan nilai di atas bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}\n({height/len(df)*100:.1f}%)',
             ha='center', va='bottom', fontsize=12, weight='bold')

plt.title('Distribusi Gender Customer', fontsize=16, weight='bold', pad=20)
plt.xlabel('Gender', fontsize=13, weight='bold')
plt.ylabel('Jumlah Customer', fontsize=13, weight='bold')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('3_gender_vertical_bar.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n3. GENDER (Vertical Bar Chart)")
print("-" * 80)
for gender, count in gender_counts.items():
    print(f"{gender}: {count} ({count/len(df)*100:.1f}%)")
```

```python
print("\nINSIGHT:")
print("- Distribusi gender cukup seimbang dengan Female 52.9% dan Male 47.1%")
print("- Tidak ada bias gender yang signifikan dalam penggunaan kartu kredit")
print("- Produk dan layanan dapat dirancang untuk kedua segmen dengan proporsi hampir sama")
```

### INSIGHT:
- Distribusi gender cukup seimbang dengan Female 52.9% dan Male 47.1%
- Tidak ada bias gender yang signifikan dalam penggunaan kartu kredit
- Produk dan layanan dapat dirancang untuk kedua segmen dengan proporsi hampir sama

## 1.4 EDUCATION LEVEL - HORIZONTAL BAR CHART

```python
plt.figure(figsize=(12, 7))
education_counts = df['Education_Level'].value_counts().sort_values()
colors_edu = plt.cm.viridis(np.linspace(0.2, 0.9, len(education_counts)))

bars = plt.barh(education_counts.index, education_counts.values, color=colors_edu,
                edgecolor='black', linewidth=1.2, alpha=0.85)

# Menambahkan nilai di ujung bar
for i, (idx, val) in enumerate(education_counts.items()):
    plt.text(val + 50, i, f'{int(val)} ({val/len(df)*100:.1f}%)',
             va='center', fontsize=11, weight='bold')

plt.title('Distribusi Tingkat Pendidikan Customer', fontsize=16, weight='bold', pad=20)
plt.xlabel('Jumlah Customer', fontsize=13, weight='bold')
plt.ylabel('Tingkat Pendidikan', fontsize=13, weight='bold')
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.savefig('4_education_horizontal_bar.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n4. EDUCATION LEVEL (Horizontal Bar Chart)")
print("-" * 80)
for edu, count in education_counts.sort_values(ascending=False).items():
    print(f"{edu}: {count} ({count/len(df)*100:.1f}%)")
```

### INSIGHT:
- Mayoritas customer adalah lulusan Graduate (30.1%) dan High School (23.2%)
- Customer dengan pendidikan tinggi (Post-Graduate, Doctorate) sekitar 18%+15.2% = 33.2%
- Tingkat pendidikan yang tinggi bisa berkorelasi dengan kemampuan finansial dan penggunaan kredit

## 1.5 INCOME CATEGORY - VERTICAL BAR CHART

```python
plt.figure(figsize=(12, 7))
income_order = ['Less than $40K', '$40K - $60K', '$60K - $80K', '$80K - $120K', '$120K +', 'Unknown']
income_counts = df['Income_Category'].value_counts()
income_counts = income_counts.reindex(income_order)

colors_income = ['#e74c3c', '#e67e22', '#f39c12', '#2ecc71', '#27ae60', '#95a5a6']
bars = plt.bar(range(len(income_counts)), income_counts.values, color=colors_income,
               edgecolor='black', linewidth=1.5, alpha=0.85)

# Menambahkan nilai di atas bar
for i, (idx, val) in enumerate(income_counts.items()):
    plt.text(i, val, f'{int(val)}\n({val/len(df)*100:.1f}%)',
             ha='center', va='bottom', fontsize=10, weight='bold')

plt.xticks(range(len(income_counts)), income_counts.index, rotation=45, ha='right')
plt.title('Distribusi Kategori Pendapatan Customer', fontsize=16, weight='bold', pad=20)
plt.xlabel('Kategori Pendapatan', fontsize=13, weight='bold')
plt.ylabel('Jumlah Customer', fontsize=13, weight='bold')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('5_income_vertical_bar.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n5. INCOME CATEGORY (Vertical Bar Chart)")
print("-" * 80)
for income, count in income_counts.items():
    print(f"{income}: {count} ({count/len(df)*100:.1f}%)")
print("\nINSIGHT:")
print("- Mayoritas customer berada di kategori pendapatan menengah ($40K-$80K) sekitar 52%")
print("- Customer berpendapatan tinggi ($120K+) hanya 7.3%, menunjukkan potensi premium segment")
print("- 11.7% customer tidak melaporkan pendapatan, perlu strategi untuk mendapatkan data lengkap")
```

### INSIGHT:
- Mayoritas customer berada di kategori pendapatan menengah ($40K-$80K) sekitar 52%
- Customer berpendapatan tinggi ($120K+) hanya 7.3%, menunjukkan potensi premium segment
- 11.7% customer tidak melaporkan pendapatan, perlu strategi untuk mendapatkan data lengkap

# Soal no. 4
SCATTERPLOT - KOMBINASI LEBIH DARI 1 KOLOM

```python
plt.figure(figsize=(14, 8))
scatter = plt.scatter(df['Customer_Age'], df['Credit_Limit'],
                      c=df['Attrition_Flag'].map({'Existing Customer': 0, 'Attrited Customer': 1}),
                      cmap='coolwarm', alpha=0.6, s=50, edgecolors='black', linewidth=0.5)

plt.colorbar(scatter, label='Attrition (0=Existing, 1=Attrited)', ticks=[0, 1])
plt.title('Hubungan Usia Customer dengan Credit Limit', fontsize=16, weight='bold', pad=20)
plt.xlabel('Usia Customer (tahun)', fontsize=13, weight='bold')
plt.ylabel('Credit Limit ($)', fontsize=13, weight='bold')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('7_scatterplot_age_credit.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n" + "=" * 80)
print("\n4. SCATTERPLOT - CUSTOMER AGE vs CREDIT LIMIT")
print("=" * 80)
```

# Soal no. 5
HEATMAP - KORELASI ANTAR KOLOM

```python
# Pilih kolom numerik untuk korelasi
numerical_cols = ['Customer_Age', 'Dependent_count', 'Months_on_book',
                  'Total_Relationship_Count', 'Months_Inactive_12_mon',
                  'Contacts_Count_12_mon', 'Credit_Limit', 'Total_Revolving_Bal',
                  'Total_Trans_Amt', 'Total_Trans_Ct']

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='RdYlGn',
            center=0, square=True, linewidths=1, cbar_kws={"shrink": 0.8},
            vmin=-1, vmax=1)

plt.title('Heatmap Korelasi Antar Variabel Numerik', fontsize=16, weight='bold', pad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('8_heatmap_correlation.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n" + "=" * 80)
print("\n5. HEATMAP - KORELASI ANTAR VARIABEL NUMERIK")
print("=" * 80)
```

3. Korelasi LEMAH/TIDAK ADA:

   - Dependent_count tidak signifikan mempengaruhi perilaku transaksi

4. IMPLIKASI BISNIS:
   - Fokus pada meningkatkan frekuensi transaksi (Total_Trans_Ct) akan meningkatkan nilai transaksi
   - Customer dengan saldo revolving tinggi perlu didorong untuk lebih aktif bertransaksi
   - Monitoring periode inactive penting untuk mencegah penurunan aktivitas transaksi
