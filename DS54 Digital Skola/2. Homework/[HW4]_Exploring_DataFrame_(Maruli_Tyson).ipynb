```python
# Import Libraries

import pandas as pd
```

```python
# Import Dataset Paid Transaction
df_paidTransaction = pd.read_excel('Homework-DataFrame-Dataset.xlsx', sheet_name="Paid-Transaction")

display(df_paidTransaction)
```

```python
# Import Dataset Seller
df_seller = pd.read_excel('Homework-DataFrame-Dataset.xlsx', sheet_name="Seller")

display(df_seller)
```

# Soal 1
Buatlah kolom baru bernama “Full Name” dan sisipkan kolom tersebut setelah kolom “Last Name” yang berisikan nama lengkap dari setiap isian record dengan ketentuan sebagai berikut:
*   Nama lengkap menggabungkan “First Name” dan “Last Name”.
*   Nama lengkap menggunakan “title case”, yaitu untuk setiap kata, huruf awalnya harus huruf kapital

```python
# Get index position of "Last Name"
new_column_index = df_paidTransaction.columns.get_loc('Last Name')

# Insert new column 'Full Name' inside the DataFrame at position after column "Last Name"
df_paidTransaction.insert(new_column_index+1, 'Full Name', (df_paidTransaction['First Name'] + ' ' + df_paidTransaction['Last Name']).str.title())

df_paidTransaction
```

# Soal 2


*   Tambahkan sebuah kolom setelah kolom “Seller Discount” dan beri nama kolom tersebut sebagai “GMV”.  
*   GMV adalah Gross Merchandise Value yang nilainya dihitung berdasarkan “Transaction Amount” dikurangi “Seller Discount” ditambah dengan “Delivery Fee”

```python
# Get index position of "Seller Discount"
new_column_index = df_paidTransaction.columns.get_loc('Seller Discount')


# Insert new column 'GMV' inside the DataFrame at position after column "Seller Discount"
df_paidTransaction.insert(new_column_index+1, 'GMV', (df_paidTransaction['Transaction Amount'] - df_paidTransaction['Seller Discount'] + df_paidTransaction['Delivery Fee']))

df_paidTransaction
```

## 3.1
Kelompokkan setiap Meta Category menjadi 2 group sesuai tabel di bawah ini (hint: buat kolom tambahan yang berisi keterangan group 1 atau 2 sesuai Meta Category-nya).

```python
# Get index position of "Meta Category"
new_column_index = df_paidTransaction.columns.get_loc('Meta Category')

group_category = {
    "Group 1": ["Home/ Food", "Sports/ Hobi/ Otomotif", "Fashion", "Beauty/ Health"],
    "Group 2": ["Gadget/ Komputer", "Elektronik", "Service/ Mokado", "Babies/ Kids"]
}

def get_group_category(item):
  for group_name, categories in group_category.items():
        if item in categories:
            return group_name
  return None

# Insert new column 'Group Meta Category' inside the DataFrame at position after column "Meta Category"
df_paidTransaction.insert(new_column_index+1, 'Group Name', df_paidTransaction['Meta Category'].map(get_group_category))

df_paidTransaction
```

## 3.2
Buatlah DataFrame baru yang menyimpan pivot table dari DataFrame “Paid-Transaction” di atas untuk menampilkan jumlah total nilai GMV per-bulan untuk setiap group.

```python
# Change column "PaidDate" to DateTime Type
df_paidTransaction['Paid Date']= pd.to_datetime(df_paidTransaction['Paid Date'], format='%Y%m%d')

df_paidTransaction
```

```python
# Create pivot table for sum of GMV for each month for each group category
pivot_gmv_group = df_paidTransaction.pivot_table(
    values='GMV',
    index=df_paidTransaction['Paid Date'].dt.month,
    columns='Group Name',
    aggfunc='sum',
    fill_value=0
).sort_index()

pivot_gmv_group.style.format('{:,.0f}')
```

# Soal 4
Buatlah python statement untuk mencari Seller mana yang memiliki total nilai GMV paling tinggi di bulan Agustus 2017. (Hint: gunakan juga sheet “Seller”)

```python
# Merge 2 DataFrame df_paidTransaction and df_Seller on "Order Number"
merged_df = pd.merge(df_paidTransaction, df_seller, on='Order Number', how='left')

merged_df
```

```python
# Filter and group by month August year 2017
august_2017 = df_paidTransaction[
    (df_paidTransaction['Paid Date'].dt.month == 8) &
    (df_paidTransaction['Paid Date'].dt.year == 2017)
]

# Sort descending to get the highest sum and get the first row
top_seller_df = august_2017.groupby(merged_df['Seller'])['GMV'].sum().sort_values(ascending=False).head(1).to_frame()

# Change format
top_seller_df.style.format({'GMV': '{:,.0f}'})
```

# Soal 5
Buatlah python statement untuk mencari Seller mana yang memiliki banyaknya transaksi (count) paling banyak di bulan September 2017 khusus untuk Meta Category “Fashion”. (Hint: gunakan juga sheet “Seller”)

```python
# Filter and group by month September year 2017 and meta category = 'Fashion'
september_2017_fashion = df_paidTransaction[
    (df_paidTransaction['Paid Date'].dt.month == 9) &
    (df_paidTransaction['Paid Date'].dt.year == 2017) &
    (df_paidTransaction['Meta Category'] == 'Fashion')
]

# Sort descending dan ambil yang teratas
top_seller_df = september_2017_fashion.groupby(merged_df['Seller']).size().sort_values(ascending=False).head(1).to_frame()

print(top_seller_df)
```
